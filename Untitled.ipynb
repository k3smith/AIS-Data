```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
import numpy as np
```

```python
world_map = gpd.read_file('WorldShp/World_Countries__Generalized_.shp')
crs = {'init':'epsg:4326'}
```

```python
file_list = ["AIS_2022_01_01.csv", 
             "AIS_2022_01_02.csv", 
             "AIS_2022_01_03.csv", 
             "AIS_2022_01_04.csv", 
             "AIS_2022_01_05.csv", 
             "AIS_2022_01_06.csv", 
             "AIS_2022_01_07.csv", 
             "AIS_2022_01_08.csv", 
             "AIS_2022_01_09.csv", 
             "AIS_2022_01_10.csv"]

df = pd.DataFrame()

for f in file_list :
    df_sub = pd.read_csv(f, usecols=['LAT', 'LON', 'CallSign', 'VesselType', 'Cargo'])
    df_sub = df_sub[df_sub['CallSign'].notna()]
    df_sub = df_sub[df_sub['VesselType'] < 90]
    df_sub = df_sub[df_sub['VesselType'] >= 70]
    df_sub = df_sub.drop_duplicates(subset=['LAT', 'LON', 'CallSign'], keep='first')
    df = pd.concat([df, df_sub])
    del df_sub
    print('Data frame now has ' + str(len(df)) + ' entries.')
    
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
world_map.plot(ax=ax)
plt.show()
```

```python
def getTop(df, col, n = 5) :
    #by_col = df[col].value_counts()
    top =  df.value_counts(col).head(n).index
    return top
```

```python
df_filtered = df[df['CallSign'].isin(getTop(df, 'CallSign', n=60))]
#df_filtered = df[df['CallSign'].map(df['CallSign'].value_counts()) >= 2]
df_filtered.head()
```

```python
df_filtered.value_counts('CallSign')
```

```python
geometry = [Point(xy) for xy in zip(df_filtered['LON'], df_filtered['LAT'])]
geo_df = gpd.GeoDataFrame(df_filtered, #specify our data
                          crs=crs, #specify our coordinate reference system
                          geometry=geometry) #specify the geometry list we created
geo_df.head()

geo_df2 = geo_df.groupby('CallSign')['geometry'].apply(lambda x: LineString(x.tolist()))
geo_df2 = gpd.GeoDataFrame(geo_df2, geometry='geometry')
```

```python
cc = pd.DataFrame([['Corpus Christi', 27.81278, -97.40722], 
                   ['New York', 40.6840, -74.0062],
                   ['Virginia', 36.9185, -76.3001],
                   ['Charleston', 32.7815, -79.9238],
                    ['Houston', 29.7300, -95.2724]], columns=['PORT', 'LAT', 'LON'])
geometry = [Point(xy) for xy in zip(cc['LON'], cc['LAT'])]
geo_port = gpd.GeoDataFrame(cc, #specify our data
                          crs=crs, #specify our coordinate reference system
                          geometry=geometry) #specify the geometry list we created
```

```python
color_labels = geo_df['CallSign'].unique()

fig, ax = plt.subplots(figsize=(15,15))
world_map.plot(ax=ax, alpha=0.4, color='grey')
#geo_df2.plot(ax=ax)

for p in color_labels :
    try :
        geo_df[geo_df['CallSign']==p].plot(ax=ax, label=p, markersize=6, marker='o', alpha=0.1)
    except :
        print(p)
geo_port.plot(ax=ax, markersize=25, marker='*')
plt.legend(prop={'size':15})
plt.show()
```
